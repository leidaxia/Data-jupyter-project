```python
g_1_list = [df[df['target'] == '是']['g'].mean(), df[df['target'] == '是']['g'].std()]
g_2_list = [df[df['target'] == '否']['g'].mean(),df[df['target'] == '否']['g'].std()]
h_1_list = [df[df['target'] == '是']['h'].mean(), df[df['target'] == '是']['h'].std()]
h_2_list = [df[df['target'] == '否']['h'].mean(), df[df['target'] == '否']['h'].std()]
```

```python
def get_gaussian(num, pro_list):
    pro_mean = pro_list[0]
    pro_std = pro_list[1]
    result = 1/(math.sqrt(2*math.pi)*pro_std)
    result = result*math.exp((-(num-pro_mean)**2)/(2*pro_std**2))
    return result
    
```

```python
#乌黑	稍蜷	浊响	清晰	稍凹	软粘	0.360	0.370	否
t_1 = a_1_dict['乌黑']*b_1_dict['稍蜷']*c_1_dict['浊响']*d_1_dict['清晰']*e_1_dict['稍凹']*f_1_dict['软粘']*get_gaussian(0.360,g_1_list)*get_gaussian(0.370,h_1_list)
t_2 = a_2_dict['乌黑']*b_2_dict['稍蜷']*c_2_dict['浊响']*d_2_dict['清晰']*e_2_dict['稍凹']*f_2_dict['软粘']*get_gaussian(0.360,g_2_list)*get_gaussian(0.370,h_2_list)
```
